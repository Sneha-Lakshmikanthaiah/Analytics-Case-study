### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the datasets *student-mat* and *student-por* and merge them.

```python
student_por = pd.read_csv("student-por.csv")
student_mat = pd.read_csv("student-mat.csv")
```

```python
df = pd.merge(left = student_por,
               right = student_mat,
               left_on = 'school',
               right_on = 'school',
               how = 'inner',
               indicator = True)

df.head()
```

### Step 3. Assign it to a variable called df.

### Step 9. Create a function called majority that return a boolean value to a new column called legal_drinker

```python
def majority(x):
    if x > 17:
        return True
    else:
        return False
```

```python
def times10(x):
    if type(x) is int:
        return 10 * x
    return x
```
