### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the  datasets [cars1.csv](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/Merge/Auto_MPG/cars1.csv) and [cars2.csv](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/Merge/Auto_MPG/cars2.csv) from the folder

```python
pd.read_csv('cars1.csv')
pd.read_csv('cars2.csv')
```

   ### Step 3. Assign each to a to a variable called cars1 and cars2

```python
cars_1 = pd.read_csv('cars1.csv')
cars_2 = pd.read_csv('cars2.csv')
```

### Step 4. It seems our first dataset has some unnamed blank columns, fix cars1

```python
cars_1.dropna(how='all', axis='columns',inplace=True)

cars_1
```

### Step 5. What is the number of observations in each dataset?

```python
print("Total number of non-null Obesrvations of Cars_1 = ",cars_1["mpg"].count())
print("Total number of non-null Obesrvations of Cars_2 = ",cars_2["mpg"].count())
```

### Step 6. Join cars1 and cars2 into a single DataFrame called cars

```python
cars = pd.concat([cars_1,cars_2],axis=0,sort=True)

cars
```

### Step 7. There is a column missing, called owners. Create a random number Series from 15,000 to 73,000.

```python
owners = np.random.randint(15000,73000,size=len(cars))
owners
```

### Step 8. Add the column owners to cars

```python
cars['Owners'] = owners
```
