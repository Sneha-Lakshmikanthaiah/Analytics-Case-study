### Step 4. Create a histogram with the 10 countries that have the most 'Quantity' ordered except UK

```python
d = online_rt.loc[online_rt['Country']!='United Kingdom']
group_country = d.groupby(['Country'])['Quantity'].sum()
largest = group_country.nlargest(10)

#graph
plt.hist(largest,color=['yellow'])
plt.xlabel('Country')
plt.ylabel('Quantity')
plt.show()
```
